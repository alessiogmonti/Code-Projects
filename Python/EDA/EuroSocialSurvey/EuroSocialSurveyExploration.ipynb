```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', 223)
pd.set_option('display.max_rows', 100)
plt.rcParams['figure.dpi'] = 300


df = pd.read_csv('ESS1-9e01_1.csv')
df.describe()
```

```python
x = df[['hinctnta', 'hincsrca', 'hincfel', 'crmvct']]

x = x[x.hinctnta < 11]
x = x[x.hincsrca < 9]
x = x[x.hincfel < 5]
x = x[x.crmvct < 3]

# x['hinctnta'].hist(bins= 10)
# plt.show()
```

```python
sns.displot(x, x="hinctnta", kind="kde", hue="hincfel", fill=False, rug=True, palette='mako')
sns.displot(x, x="hincsrca", kind="kde", hue="hincfel", fill=False, rug=True, palette='mako')
```

```python
sns.displot(x, x="hinctnta", y="hincsrca", kind="kde", hue="hincfel", fill=True, rug=True, palette='mako')
```

```python
sns.heatmap(data=x.corr(), cmap='vlag')
```

```python
sns.displot(x, x="hinctnta", y="hincfel", hue="hincsrca", kind="kde", fill=True, rug=True, palette='mako')
```

```python
sns.jointplot(x=x['hinctnta'], y=x['hincfel'], kind="hex", color="#4CB391")
```

```python
sns.violinplot(x=x['hincsrca'], y=x['hinctnta'], hue=x["crmvct"], split=True, alpha=0.1, palette="BuGn")
```

```python
sns.violinplot(x=x['hincsrca'], y=x['hinctnta'], alpha=0.1, palette="BuGn", scale="count")
```

```python
sns.violinplot(x=x['hincfel'], y=x['hinctnta'], hue=x["crmvct"], split=True, alpha=0.1, palette="BuGn", scale="count")
```
