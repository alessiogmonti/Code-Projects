```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import norm
from scipy import stats
from sklearn.preprocessing import StandardScaler

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
sns.set(style='white', context='notebook', rc={'figure.figsize':(8,8),'figure.dpi':(300)})
```

```python
df_train = pd.read_csv('house-prices-advanced-regression-techniques/train.csv')
display(df_train.head())
```

```python
sns.distplot(df_train['SalePrice'])
```

```python
print("Skewness: %f" % df_train['SalePrice'].skew()) #measure of the asymmetry of the data's probability distribution compared to a normal distribution
print("Kurtosis: %f" % df_train['SalePrice'].kurt()) #degree to which the data is distributed at the tails
```

```python
sns.scatterplot(x=df_train['SalePrice'], y=df_train['GrLivArea'])
```

```python
sns.boxplot(x=df_train['OverallQual'], y=df_train['SalePrice'], palette=sns.light_palette('purple'))
```

```python
f, ax = plt.subplots(figsize=(20, 8))
fig = sns.boxplot(x=df_train['YearBuilt'], y=df_train["SalePrice"], palette='Greens_r')
fig.axis(ymin=0, ymax=800000);
plt.xticks(rotation=90);
```

```python
corrmat = df_train.corr()
f, ax = plt.subplots(figsize=(9, 9))
sns.heatmap(corrmat, vmax=.9, square=True, linewidths=.5, cmap="BuPu");
```

```python
k = 10 #number of variables for heatmap
cols = corrmat.nlargest(k, 'SalePrice')['SalePrice'].index
cm = np.corrcoef(df_train[cols].values.T)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values, linewidths=.5, cmap="BuPu")
plt.show()
```
