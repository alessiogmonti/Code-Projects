```python
import pandas as pd
import pandas_datareader as web
import datetime as dt
import seaborn as sns
import matplotlib.pyplot as plt
import talib
```

```python
sns.set(style='white', context='paper', rc={'figure.figsize':(10,5),'figure.dpi':(1200)})
```

```python
tickers = ['BABA', 'CRWD', 'PLTR', 'SPCE', 'TSLA']

start = '09-30-2020'
end = dt.date.today()

df = web.DataReader(tickers,'yahoo', start, end)
df.dtypes
```

```python
display(df)
```

```python
adjclose = {}
for x in tickers:
    adjclose[x] = df['Adj Close'][x]
```

```python
pctClose = {}
for k,v in adjclose.items():
    pctClose[k] = pd.DataFrame(v.pct_change()).dropna()
```

```python
for x in pctClose:
    sns.lineplot(data=pctClose[x])
plt.show()
```
