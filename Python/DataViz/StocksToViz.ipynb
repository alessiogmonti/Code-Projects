```python
import pandas as pd
import pandas_datareader as web
import datetime as dt
import seaborn as sns
import talib
```

```python
sns.set(style='white', context='paper', rc={'figure.figsize':(10,5),'figure.dpi':(1200)})
```

```python
start = '09-30-2020'
end = dt.date.today()

df = web.DataReader('PLTR','yahoo', start, end)
df.dtypes
```

```python
sns.lineplot(data=df, y='Volume', x=df.index, color='Seagreen')
```

```python
close = df['Adj Close']
high = df['High']
low = df['Low']

upperband, middleband, lowerband = talib.BBANDS(close, timeperiod=5, nbdevup=2, nbdevdn=2, matype=0)
ema = talib.EMA(close, timeperiod=20)
mama, fama = talib.MAMA(close, fastlimit=0.1, slowlimit=0.2)
tema = talib.TEMA(close, timeperiod=30)

res = [upperband,middleband,lowerband,ema,mama,fama,tema]
res_names = ['upperband', 'middleband', 'lowerband', 'ema', 'mama', 'fama', 'tema']

tf = pd.concat(res, axis=1, keys=res_names)
df = pd.concat([close,tf],axis=1)
```

```python
pct = df['Adj Close'][-50:].pct_change()
```

```python
sns.lineplot(data=pct)
```

```python
start = '09-30-2020'
end = dt.date.today()

tickers = ['AMZN', 'AAPL', 'TSLA', 'PLTR', 'DAL']

df1 = web.DataReader(tickers,'yahoo', start, end)['Adj Close']
```

```python
df1['ord'] = df1.index
df1['ord'] = df1.ord.apply(lambda date: date.toordinal())
```

```python
import matplotlib.pyplot as plt

run = False
if run:
    for j in df1.columns[:-1]:
        sns.kdeplot(data=df1, x=df1.ord, y=j,fill=False, color='seagreen')
        sns.lineplot(data=df1, x=df1.ord, y=j, linewidth=0.2, color='black')
        ax = sns.regplot(data=df1, x=df1.ord, y=j, marker='s', scatter_kws={'s':2, },color='grey', line_kws={'linewidth':0.1})
        ax.set_xlabel('date')
        new_labels = [dt.date.fromordinal(int(item)) for item in ax.get_xticks()]
        ax.set_xticklabels(new_labels)
        plt.savefig("test{}.svg".format(j), format="svg")
        plt.show()
```
